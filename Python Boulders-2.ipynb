```python
"""
By: Yvette Yi Yuan
"""

"""
Docstring:

    A). Introduction:
    ***************************************************************
    If you haven’t climbed, do you really do tech in silicon valley?
    ***************************************************************
    Climbing is about building camaraderie, overcoming physical obstacles and 
    frustrations, and taking on new adventures. 
    That being said, it is a dangerous sport! 
    
    In this simulated climbing game, you will make decisions about choosing 
    the right gears, preparing for climbing and to climb with style in an 
    indoor climbing gym. You will learn safety tips and fundamental techniques 
    for bouldering*, so you are better prepared to enjoy climbing safely. 
    
    Round A: Choose gears
    Round B: Pre-climbing
    Round C: Climbing
    Round D: Falling from the wall
      
    If you are already a seasoned climber, enjoy a good send*! 
    
    Bouldering*: a type of climbing, typically on small rock formations or 
    short walls without ropes or harnesses.
    Send*: climbing a route without falling.
      
    B). Error:
    when calling exit(0):
    An exception has occurred, use %tb to see the full traceback.
    SystemExit: 0
    """
#importing in the global environment
from time import sleep 
from random import randint
from sys import exit

def start_game():
    #input user name and acknowledge climbing is dangerous 
    global first_name
    first_name = input("""
        Climbling is an inherently dangerous sport. 
        Before we begin, please sign your name at the bottom of this waiver to 
        declare you are fully aware of the potential harms and that you will be 
        fully responsible for your own safety.\n \n\tFirst name > """)

    first_name = first_name.capitalize()
    last_name = input("\tLast name >")

    print(f"""
        Welcome to the Python Boulders, {first_name}!""")
    gears()

#choosing shoes and other gears
def gears():
    global toolkit
    toolkit = []
    shoe_size = input("""
    You are at the gears rental booth. What size of shoes do you typically 
    wear? \n\t> """)
    
    #control for error
    while True:
        try: 
            shoe_size = int(shoe_size)
            break
        except ValueError: 
            shoe_size = input("\tOops, that is not a size. What size do you wear? \n\t>")
    shoe_size = int(shoe_size)
    climbing_shoe_size = shoe_size - 1
    print(f"""
    We recommend that you go one size down to {climbing_shoe_size}. It is 
    uncomfortable, but tighter fit helps you make precise moves.\n """)
    input("\t<Press enter to put on your climbing shoes>")
    
    #choose more gears
    print("""
    We have more goodies for you! Pick some to take with you.
    1. chalk 
    2. shirtlessness
    3. low ponytail for men
    4. protein bar""")
    more_gears = input("\tChoose the gears you want \n\t>")
    more_gears = more_gears.casefold()
    
    #1&2&3, especially 2&3 increases your chance of winning. 
    if "1" in more_gears: 
        toolkit.append("chalk")
    if "chalk" in more_gears: 
        toolkit.append("chalk")
    if "2" in more_gears: 
        toolkit.append("shirtless")
    if "shirtless" in more_gears: 
        toolkit.append("shirtless")
    if "3" in more_gears: 
        toolkit.append("ponytail")
    if "ponytail" in more_gears: 
        toolkit.append("ponytail")
    if "4" in more_gears:
        toolkit.append("protein bar")
    if "bar" in more_gears: 
        toolkit.append("protein bar")
    #proceed to warm_up    
    print(f"""\tYou have {toolkit} with you.""")
    warm_up()

def warm_up():
    print("""
    Would you like to do some warm-up exercise? Y/N\n""")
    answer = input("\t>")
    answer = answer.casefold()

    while True:
        if "n" in answer:
            climb()
        elif "y" in answer:
            #warmup prevents injury(), which leads to end of game
            toolkit.append("warmup")
            print("\tJumping jack")
            sleep(1)
            print("\tMountain climb")
            sleep(1)
            print("\tJumping jack")
            sleep(1)
            print("\tMountain climb")
            sleep(1)        
            print("\tJumping jack")
            sleep(1)
            break
        #control for error
        else:
            answer = input("\tAhem, yes or no to warm-up?\t>")
    climb()

#3 moves required to complete the climb, energy >=2 is also required      
def climb():
    #importing in the local environment
    number_moves = 3
    energy = 3
    print("Now we are at foot the bouldering wall, ready to climb! ")
    input("Press ENTER to begin climbing >\n")
    print("You are on the bouldering wall. ")
    while number_moves > 0:
        #generates random movement descriptions
        angle1 = randint(0,90)
        angle2 = randint(0,90)
        direction_lst = ["left", "right"]
        direction1 = direction_lst[randint(0,1)]
        direction2 = direction_lst[randint(0,1)]
        technique = input(f"""You have a few options... How do you proceed?\n 
        1. Reach for the hold {angle1} degrees upper {direction1} with your hand.
        2. Reach for the hold {angle2} degrees lower {direction2} with your foot. 
        >""")
        number_moves -= 1
        
        #Reaching for footholds is a good habit that prevents reducing energy level by 1; 
        #Reaching for handholds is a bad habit that quickly exhausts the climber, 
            #reducing energy level by 1
        if technique == "1":
            energy -= 1
            print("There you go! ")
            sleep(0.5)
        elif technique == "2":
            print("Good technique! ")
            sleep(0.5)
        #control for error
        else:
            technique = input("\tI mean, what's your next move, 1 or 2?\t>")           
        
        #If you are shirtless or wear a man ponytail, you must be a pro-climber.
        while energy <= 2: 
            if "shirtless" in toolkit:
                break
            elif "ponytail" in toolkit:
                break
            else: 
                print("Oh no! Your forearms are pumped, you are falling...")
                fall()         
        
        #Not having warmup in the toolkit leads to injury(), end of game
        while number_moves == 2:
            if "warmup" in toolkit:
                break
            else:
                technique = input(f"""You have a few options... How do you proceed?\n 
        1. Reach for the hold {angle1} degrees upper {direction1} with your hand.
        2. Reach for the hold {angle2} degrees lower {direction2} with your foot. 
        >""")
                injury()
        
        #completes climb
        while number_moves == 0:
            print("""
                Congratulations on your send at Python Boulders!""")
            if "shirtless" in toolkit:
                print(f"""
                {first_name}, I see that you are shirtless. 
                You must be a pro-climber!""")
            elif "ponytail" in toolkit:
                print(f"""
                {first_name}, I see that you wear a men ponytail. 
                You must be a pro-climber!""")
            else: 
                print("""
                As you can see, using your feet and legs, not just pulling 
                with your arms is an awesome climbing technique! """)
            input("Here is a little something for you...\nPress ENTER >")
            beer()

#goodbye message
def bye():
    print("""
        __                  __             
       / /_  __  _____     / /_  __  _____ 
      / __ \/ / / / _ \   / __ \/ / / / _ \
     / /_/ / /_/ /  __/  / /_/ / /_/ /  __/
    /_.___/\__, /\___/  /_.___/\__, /\___/ 
          /____/              /____/       
    """)
    replay()

#prize
def beer():
    print(f"""
    Climbing is all about the post-climb beers that last late into the night!""") 
    sleep(1)
    print(f"""
    Here is a very hoppy IPA for you. 
    Cheers, {first_name}.
    
    .~~~~.
    i====i_
    |cccc|_)
    |cccc|      
    `-==-'""")
    replay()

fall_times = 0
#climb() again
def fall():
    falling = "falling"
    for letter in falling: 
        print(letter)
    global fall_times
    fall_times += 1
    print(fall_times)   
    #set limit for number of times player can climb again
    #chalk in toolkit increases allowance for another climb
    while fall_times >= 2:
        #4 tries exhausted, leads to beer()
        if fall_times == 4: 
            print(f"Thank you for trying, {first_name}! \nFalling and presevering are as much a part of climbing as getting to the top.")
            input("Here is a little something for you...\nPress ENTER >")
            beer()  
        elif "chalk" in toolkit:
            input("""Time to apply chalk! \nPress ENTER to chalk up \t>""")
            break
        #end of game
        else:
            print(f"""
            Your hands are too tired to grip, {first_name}.
            Remember to take some chalk with you next time.""")
            bye()   
    
    #offer to replay
    print("Would you like to give this route another try? Y/N")
    answer = input("\t>")
    answer = answer.casefold()
    
    while True:
        if "y" in answer:    
            climb()
        #end of game
        elif "n" in answer:
            input("\tPress ENTER to exit\t>")
            bye()
        #control for error
        else:
            answer = input("\tAnswer Y/N to continue climbing or quit.\t>")

#fail(), end of game
def injury():
    print(f"""
    
      ▄• ▄▌ ▄▄·  ▄ .▄
▪     █▪██▌▐█ ▌▪██▪▐█
 ▄█▀▄ █▌▐█▌██ ▄▄██▀▐█
▐█▌.▐▌▐█▄█▌▐███▌██▌▐▀
 ▀█▄▀▪ ▀▀▀ ·▀▀▀ ▀▀▀ ·
That hurts! Sorry that you pulled your tendon, and tendon injuries typically 
take up to 6 months to heal... 
Remember to warmup before climbing! Hope you feel better soon, {first_name}. )""")
    input("Press ENTER to exit\t>")
    bye()

#play again from the beginning
def replay():
    #importing in the local environment
    print("Would you like to play Python Boulders again? Y/N")
    answer = input("\t>")
    answer = answer.casefold()
    
    while True:
        if "y" in answer:    
            global fall_times
            fall_times = 0
            start_game()
        #end of game
        elif "n" in answer:
            input("\tPress ENTER to exit\t>")
            break
        #control for error
        else:
            answer = input("\tAnswer Y/N to continue climbing or quit.\t>")
    exit(1)

start_game()
```
